```
%%sql -r dataframe_2
CREATE VIEW DimCustomer_analytical AS
SELECT
    "customer_id",
    REGEXP_REPLACE("full_name", '(\\w)\\w+\\s(\\w)\\w+', '\\1*** \\2***') AS full_name,
    REGEXP_REPLACE("email", '(^.)(.*)(@.*)', '\\1*******\\3')              AS email,
    REGEXP_REPLACE("phone", '(\\+?\\d{2})\\d+(\\d{3})', '\\1*******\\2')  AS phone,
    "region_id",
    "segment_id",
    "signup_date",
    "gender",
    "created_at",
    "updated_at"
FROM silver.Dim_Customers;
```

```
%%sql -r dataframe_3
select * from DimCustomer_analytical
```

```
%%sql -r dataframe_4
CREATE VIEW DimDriver_analytical AS
SELECT
    "driver_id",
    REGEXP_REPLACE("driver_name", '(\\w)\\w+', '\\1***')                      AS driver_name,
    REGEXP_REPLACE("driver_phone", '(\\+?\\d{2})\\d+(\\d{3})', '\\1*******\\2') AS driver_phone,
    MD5("national_id")                                                          AS national_id,
    "region_id",
    "shift",
    "vehicle_type",
    "hire_date",
    "rating_avg",
    "on_time_rate",
    "cancel_rate",
    "completed_deliveries",
    "is_active",
    "created_at",
    "updated_at"
FROM silver.Dim_Drivers;
```

```
%%sql -r dataframe_5
select * from DimDriver_analytical
```

```
%%sql -r dataframe_6
CREATE OR REPLACE VIEW DimAgent_analytical AS
SELECT
    "agent_id",
    REGEXP_REPLACE("agent_name", '(\\w)\\w+', '\\1***')                       AS agent_name,
    REGEXP_REPLACE("agent_email", '(^.)(.*)(@.*)', '\\1*******\\3')            AS agent_email,
    REGEXP_REPLACE("agent_phone", '(\\+?\\d{2})\\d+(\\d{3})', '\\1*******\\2') AS agent_phone,
    "team_id",
    "skill_level",
    "hire_date",
    "avg_handle_time_min",
    "resolution_rate",
    "csat_score",
    "is_active",
    "created_at",
    "updated_at"
FROM silver.Dim_Agents;
```

```
%%sql -r dataframe_7
select * from DimAgent_analytical
```
